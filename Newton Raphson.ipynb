# Newton Raphson in one dimension
In this project, I write my own version of Newton-Ralphson's optimization in one-dimension. While existing libraries, such as scipy, already has optimize as a method that is readily avaliable, it is good practice to write my own version to understand the concept of it.

```python
# import dependencies
import pandas as pd
```

## Input Functions and parameters
The following includes f(x), f'(x), and the starting point X

```python
def f (x) :
  return x**6/6 - 3*x**4 - 2*x**3/3 + 27*x**2/2 + 18*x - 30

def d_f (x) :
  return x**5 - 12*x**3 - 2*x**2 + 27*x + 18

x = 1.99
```

## Newton-Raphson in one dimesnion
The formula for Newton-Raphson is:
Xi+1 = X0 - f(X0)/f'(X0)

```python
d = {"x":[x], "f(x)":[f(x)]}
for i in range(0,20):
    x = x - f(x)/d_f(x)
    d["x"].append(x)
    d["f(x)"].append(f(x))

pd.DataFrame(d, columns=["x", "f(x)"])
```
